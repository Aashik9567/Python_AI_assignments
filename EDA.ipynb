```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/sample_data.csv")
df.head(25)
```

```python
# Categorical data
# countplot
counts = df['ProductCategory'].value_counts()
ax=counts.plot(kind='bar', rot=0, color='darkblue', figsize=(8, 6), title='Count Plot of ProductCategories')
ax.set_xlabel('ProductCategories')
ax.set_ylabel('Count')
plt.show()
```

```python
 # pie chart

df['ProductBrand'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df['CustomerAge'])
```

```python
sns.distplot(df['CustomerAge'],color='orange')
plt.xlabel('Customer age')
plt.ylabel('Density')
plt.title('Distribution Plot')
```

```python
sns.boxplot(df['CustomerAge'])
```

```python
sns.barplot(x='ProductBrand',y='CustomerAge',hue='ProductBrand',data=df)
```

```python
sns.boxplot(x='ProductBrand',y='CustomerAge',hue='CustomerGender',data=df)
```

```python
sns.distplot(df[df['CustomerGender']==0]['CustomerAge'],hist=False)
sns.distplot(df[df['CustomerGender']==1]['CustomerAge'],hist=False)
```

```python
sns.heatmap(pd.crosstab(df['ProductCategory'],df['ProductBrand']))
```

```python
sns.pairplot(df.head(25),hue='ProductBrand')
```

```python
sns.lineplot(df.head(25),x='CustomerAge',y='ProductBrand')
```

```python
sns.heatmap(df.pivot_table(values='CustomerAge',index='CustomerGender',columns='CustomerSatisfaction'))
```
